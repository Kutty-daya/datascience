```python
list1 = list(map(int,input("Enter a list").split()))
list2 = list(map(int,input("Enter a list").split()))
print("l1 : ",list1)
print("l2 : ",list2)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))
plt.plot(list1,list2)
plt.xlabel("day")
plt.ylabel("time")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.scatter(list1, list2)
plt.xlabel("Day")
plt.ylabel("Time")
plt.title("Scatter Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.bar(list1, list2)
plt.xlabel("Day")
plt.ylabel("Time")
plt.title("Bar Chart")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.barh(list1, list2)
plt.xlabel("Time")
plt.ylabel("Day")
plt.title("Horizontal Bar Chart")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.hist(list2, bins=10)
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,8))
plt.pie(list2, labels=list1, autopct="%0.1f%%")
plt.title("Pie Chart")
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(3, 1, figsize=(10,8), sharex=True)

axs[0].plot(list1, list2)
axs[0].set_title("Plot 1")

axs[1].scatter(list1, list2)
axs[1].set_title("Plot 2")

axs[2].bar(list1, list2)
axs[2].set_title("Plot 3")

plt.xlabel("Day")
plt.tight_layout()
plt.show()
```
